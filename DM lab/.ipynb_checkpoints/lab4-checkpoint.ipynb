# Cafe Sales Data Cleaning and Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('dirty_cafe_sales.csv')

# Standardize missing values
df.replace(['UNKNOWN', 'ERROR', ' '], np.nan, inplace=True)
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

### 1. Count the number of null entries for every attribute

```python
null_counts = df.isnull().sum()
print("Null entries per attribute:\n", null_counts)
```

### 2. Fill null entries with appropriate decisions

```python
# a) if quantity is blank then fill it using 'total spent' and 'cost per unit'
df['Quantity'] = df['Quantity'].fillna(df['Total Spent'] / df['Price Per Unit'])

# b) if item is blank then fill it using cost per unit 
# We define a mapping for common prices to items based on existing frequencies
price_to_item = {2.0: 'Coffee', 3.0: 'Cake', 1.0: 'Cookie', 5.0: 'Salad', 4.0: 'Sandwich', 1.5: 'Tea'}
df['Item'] = df['Item'].fillna(df['Price Per Unit'].map(price_to_item))

# c) if the transaction date is blank then see if this can be filled by multi column sorting of transaction id
df = df.sort_values(by='Transaction ID')
# UPDATED LINE TO FIX WARNING:
df['Transaction Date'] = df['Transaction Date'].ffill().bfill()

# d) try to fill payment mode
mode_payment = df['Payment Method'].mode()[0]
df['Payment Method'] = df['Payment Method'].fillna(mode_payment)

print("Remaining nulls:\n", df.isnull().sum())
```

### 3. Split the transaction date to 3 columns as date, month and year

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df['Day'] = df['Transaction Date'].dt.day
df['Month'] = df['Transaction Date'].dt.month
df['Year'] = df['Transaction Date'].dt.year
df.head()
```

### 4. Aggregate sales based on day, month and year

```python
daily_sales = df.groupby(['Year', 'Month', 'Day'])['Total Spent'].sum().reset_index()
print("Daily Sales Aggregation:\n", daily_sales.head())
```

### 5. Aggregate the query 4 based on products

```python
product_sales = df.groupby(['Item', 'Year', 'Month', 'Day'])['Total Spent'].sum().reset_index()
print("Product Daily Sales Aggregation:\n", product_sales.head())
```

### 6. Visualize the query 4 and 5

```python
plt.figure(figsize=(12, 5))
plt.plot(daily_sales.index, daily_sales['Total Spent'])
plt.title('Overall Sales Trend')
plt.xlabel('Days (Index)')
plt.ylabel('Total Spent')
plt.show()

plt.figure(figsize=(12, 5))
# UPDATED LINE TO FIX WARNING: replaced ci=None with errorbar=None
sns.lineplot(data=product_sales, x='Day', y='Total Spent', hue='Item', errorbar=None)
plt.title('Sales Trend by Product')
plt.show()
```

### 7. Create a dashboard for the same

```python
import ipywidgets as widgets
from IPython.display import display

def plot_dashboard(item_filter):
    plt.figure(figsize=(10, 4))
    if item_filter == 'All':
        data = daily_sales
        plt.plot(data.index, data['Total Spent'])
    else:
        data = product_sales[product_sales['Item'] == item_filter]
        plt.plot(data['Day'], data['Total Spent'])
    plt.title(f'Sales for {item_filter}')
    plt.ylabel('Total Spent')
    plt.show()

items = ['All'] + list(df['Item'].dropna().unique())
dropdown = widgets.Dropdown(options=items, description='Select Item:')
widgets.interactive(plot_dashboard, item_filter=dropdown)
```
