## Task 1: Combine Files, Remove Duplicates & Format Issues
We will merge the two datasets, drop duplicates, and standardize all column names.

```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
df1 = pd.read_excel('inconsistency_B1.xlsx')
df2 = pd.read_excel('inconsistency_b2.xlsx')
df = pd.concat([df1, df2], ignore_index=True)
df.drop_duplicates(inplace=True)
cols = pd.Series(df.columns.str.strip().str.lower().str.replace(' ', '_'))
for dup in cols[cols.duplicated()].unique():
    cols[cols[cols == dup].index.values.tolist()] = [
        dup + '_' + str(i) if i != 0 else dup for i in range(sum(cols == dup))
    ]
df.columns = cols
if 'transaction_id_1' in df.columns:
    df.drop(columns=['transaction_id_1'], inplace=True)
print(f"Dataset shape after merging and deduplication: {df.shape}")
display(df.head(2))
```

## Task 2: Item Price Consistency
Each item cannot have multiple prices. We will find the most frequent price (mode) for each item and apply it universally.

```python
# Convert price to numeric, turning errors into NaNs
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')

# Find the most frequent price (mode) for each item to establish a single consistent price
mode_prices = df.groupby('item')['price_per_unit'].apply(
    lambda x: x.mode()[0] if not x.mode().empty else np.nan
).to_dict()

# Map the consistent price back to each item
df['price_per_unit'] = df.apply(
    lambda row: mode_prices.get(row['item'], row['price_per_unit']), axis=1
)

print("Item prices have been made consistent.")
```

## Task 3: Check and Fix Area Codes & Locations
Standardizing Mumbai locations and pincodes (which start with 400 or 401).

```python
# Standardize location text formatting
df['location'] = df['location'].astype(str).str.strip().str.title()

# Fix common spelling mistakes in Mumbai areas found in the data
location_fixes = {
    'Anderi': 'Andheri',
    'Mazgoan': 'Mazgaon',
    'Bandra(W)': 'Bandra West',
    'Cumbala Hills': 'Cumballa Hill',
    'Amareica': np.nan # Often accidentally entered from the card_type column
}
df['location'] = df['location'].replace(location_fixes)

# Clean pincodes (extract just the numbers, removing accidental letters/spaces)
df['pincode'] = df['pincode'].astype(str).str.extract(r'(\d+)')

# Group by location and find the correct 6-digit mode pincode (starting with 40)
valid_pincodes = df[df['pincode'].str.match(r'^40[01]\d{3}$', na=False)]
mode_pincodes = valid_pincodes.groupby('location')['pincode'].apply(
    lambda x: x.mode()[0] if not x.mode().empty else np.nan
).to_dict()

# Map the correct, consistent pincode back to the respective locations
df['pincode'] = df['location'].map(mode_pincodes).fillna(df['pincode'])

print("Locations and Area Codes standardized.")
```

## Task 4: Total Amount Spent Consistency
Total amount spent cannot be negative. We will resolve this by applying an absolute value function.

```python
# Convert total spent to numeric
df['total_spent'] = pd.to_numeric(df['total_spent'], errors='coerce')

# Convert negative values to their absolute positive equivalents
df['total_spent'] = df['total_spent'].abs()

# (Optional constraint safeguard) Recalculate if it completely mismatches price * quantity
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['calculated_total'] = df['price_per_unit'] * df['quantity']
df['total_spent'] = df['total_spent'].fillna(df['calculated_total'])
df.drop(columns=['calculated_total'], inplace=True)

print("Negative values in total spent corrected.")
```

## Task 5: Standardize Credit Card Network Providers
Permitted values are: 'VISA', 'Master card', 'American Express', 'Discover'.

```python
# Standardize formatting for mapping
df['card_type'] = df['card_type'].astype(str).str.strip().str.title()

# Dictionary to fix spelling and standardize to the 4 approved networks
card_mapping = {
    'Visa': 'VISA',
    'Mastercard': 'Master card',
    'Master Card': 'Master card',
    'Master': 'Master card',
    'American Express': 'American Express',
    'Amareica': 'American Express', # Fixing the typo found in snippet
    'Amex': 'American Express',
    'Discover': 'Discover',
    'Nan': np.nan 
}

# Apply the mapping
df['card_type'] = df['card_type'].map(card_mapping).fillna(df['card_type'])

print("Credit Card network providers standardized.")
```

## Task 6: Payment Options and UPI Sub-types
Primary payment options: 'credit card', 'cash', 'UPI'. 
If UPI, it can further be sub-typed to 'Gpay', 'bheempay', 'amazon pay', 'rupay', or 'paytm'.

```python
df['payment_method'] = df['payment_method'].astype(str).str.strip().str.lower()

# Define the categories
upi_types = ['gpay', 'google pay', 'bheema', 'bheempay', 'amazon pay', 'rupay', 'paytm', 'upi']
credit_card_types = ['credit card', 'cc']
cash_types = ['cash']

def standardize_payment(method):
    if any(u in method for u in upi_types):
        return 'UPI'
    elif method in credit_card_types:
        return 'Credit Card'
    elif method in cash_types:
        return 'Cash'
    elif method == 'nan':
        return np.nan
    else:
        return method

# Apply the categorization function
df['payment_method'] = df['payment_method'].apply(standardize_payment)

print("Payment methods standardized to Credit Card, Cash, and UPI.")
```

## Task 7: Dynamic Discount Slabs
Recalculate `discount_given` based on the spend amount slabs: e.g. 0-50 -> 3, 51-100 -> 5, 101-200 -> 10, >200 -> 15.

```python
# Define slabs (e.g., 0-50: 3%, 50-100: 5%, 100-200: 10%, >200: 15%)
# Note: Adjust the `bins` and `labels` lists if your assignment specifies different slab boundaries.
bins = [0, 50, 100, 200, np.inf]
discount_percentages = [3, 5, 10, 15] 

# Apply discounts dynamically based on the total_spent column
df['discount_given'] = pd.cut(df['total_spent'], bins=bins, labels=discount_percentages, right=True)

# Convert to standard numerical type
df['discount_given'] = pd.to_numeric(df['discount_given'])

print("Discounts recalculated based on spending slabs.")
```

## Task 8: Standardize Item Names
All item names must begin with `ITEM_0`. We will extract the number segment and format it properly.

```python
def format_item_name(item_val):
    # Safely handle missing values (NaN) or empty entries
    if pd.isna(item_val) or str(item_val).strip().lower() == 'nan' or not str(item_val).strip():
        return np.nan
    
    # Force conversion to string to prevent the TypeError
    item_str = str(item_val).strip()
    
    # Extract just the digits from messy strings (like 'Item_1_PAT', 'it,07')
    digits = re.findall(r'\d+', item_str)
    
    if digits:
        # Pad with 0 so "1" becomes "ITEM_01", "7" becomes "ITEM_07"
        num = int(digits[0])
        return f"ITEM_0{num}"
        
    return item_str

# Apply formatting
df['item'] = df['item'].apply(format_item_name)

print("Item names reformatted to standard ITEM_0X structure.")
display(df[['item']].head(10)) # Displaying the first few rows to verify the output
```

## Task 9: Check Spellings & Final Consistencies
Final pass to clean up category spelling errors (e.g. 'Pastery' -> 'Pastry'), trim strings, and remove fully corrupted empty rows.

```python
# Fix categorical spelling issues (e.g., 'Pastery' -> 'Pastry')
df['category'] = df['category'].astype(str).str.strip().str.title()

category_fixes = {
    'Pastery': 'Pastry', 
    'Nan': np.nan
}
df['category'] = df['category'].replace(category_fixes)

# Clean up empty strings that were mapped to 'Nan'
df.replace('Nan', np.nan, inplace=True)

# Drop rows where all elements are entirely missing 
df.dropna(how='all', inplace=True)

# Final reset of the index to ensure clean numbering
df.reset_index(drop=True, inplace=True)

print("Final cleanup complete. The dataset is consistent.")
display(df.head())
```
